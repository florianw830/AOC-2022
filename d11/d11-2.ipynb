```python
#from https://trainyourprogrammer.de/python-116-bestimmung-des-kleinsten-gemeinsamen-vielfachen-kgv.html
def kgV(m,n):
    """ calculate the least common multiple """
    if m == 0 or n == 0:
        return(0)
    else:
        if m > n:
            m,n = n,m
        mCounter = m;
        while mCounter <= m*n:
            if mCounter % n == 0:
                return(mCounter)
                break
            mCounter += m
```

```python
class Testset():
    def __init__(self,md,t,f,kgv):
        self.md = md
        self.t = t
        self.f = f
        self.retT = None
        self.retF = None
        self.kgv = kgv
    def runtest(self,x):
        if x % self.md ==0:
            return self.retT,x % self.kgv
        return self.retF,x % self.kgv

class Item():
    def __init__(self,value):
        self.oldValues = []
        self.value = value
    def setValue(self,v):
        self.oldValues.append(self.value)
        self.value = v
    def __str__(self) -> str:
        return "H" + str(self.value)+','.join( [str(x) for x in self.oldValues] )+ "|"
    

class Monkey():
    def __init__(self,items, no, testset,op,opList):
        self.items = items
        self.no = no
        self.testset = testset
        self.op = op
        self.opList = opList
        self.opValue = 0
        self.inspeted = 0

    def operation(self,x):
        if self.opList[0] =="old" and self.opList[1]=="old" :
            return self.op(x,x)
        return self.op(x,int(self.opList[1]))
    def getList(self):
        return [str(x) for x in self.items]
    def inspect(self):
        rmvList = []
        for itm in self.items:
            self.inspeted+=1
            v = self.operation(itm.value)

            tar,v = self.testset.runtest(v)
            itm.setValue(v)

            tar.items.append(itm)
            rmvList.append(itm)
        for itm in rmvList:
            self.items.remove(itm)
            
    def __lt__(self,other):
        return self.inspeted < other.inspeted



        
    @staticmethod
    def fromList(s,kgv):

        no = s[0].split(" ")[1][:-1]
        itms = [Item(int(x.strip())) for x in s[1].split(":")[1].split(",")]

        o1 = s[2].split(":")[1].split(" ")[3]
        op = s[2].split(":")[1].split(" ")[4]
        o2 = s[2].split(":")[1].split(" ")[5]
        test = int(s[3].split(" ")[-1])
        t = int(s[4].split(" ")[-1])
        f = int(s[5].split(" ")[-1])
        l = None
        if op == "+":
            l = lambda x,y : x+y
        if op == "*":
            l = lambda x,y : x*y
        opList = [o1,o2]
        testset = Testset(test,t,f,kgv)
        print(no)
        ret = Monkey(itms,no,testset,l,opList)
        return ret






```
