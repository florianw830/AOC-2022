```python
f = open("input.txt", "r")
l = f.read().split("\n\n")

ll = [x.split("\n") for x in l]
```

```python
class Testset():
    def __init__(self,md,t,f):
        self.md = md
        self.t = t
        self.f = f
        self.retT = None
        self.retF = None
    def runtest(self,x):
        if x % self.md ==0:
            return self.retT
        return self.retF

class Item():
    def __init__(self,value):
        self.value = value
    def __str__(self) -> str:
        return self.value
    

class Monkey():
    def __init__(self,items, no, testset,op,opList):
        self.items = items
        self.no = no
        self.testset = testset
        self.op = op
        self.opList = opList
        self.opValue = 0
        self.inspeted = 0
    def operation(self,x):
        if self.opList[0] =="old" and self.opList[1]=="old" :
            return self.op(x,x)
        return self.op(x,int(self.opList[1]))
    def getList(self):
        return [x.value for x in self.items]
    def inspect(self):
        rmvList = []
        for itm in self.items:
            self.inspeted+=1
            v = self.operation(itm.value)//3
            
            tar = self.testset.runtest(v)
            itm.value = v
            print(self.no,"throw", itm.value,"to", tar.no)
            tar.items.append(itm)
            rmvList.append(itm)
        for itm in rmvList:
            self.items.remove(itm)
            
    def __lt__(self,other):
        return self.inspeted < other.inspeted



        
    @staticmethod
    def fromList(s):

        no = s[0].split(" ")[1][:-1]
        itms = [Item(int(x.strip())) for x in s[1].split(":")[1].split(",")]

        o1 = s[2].split(":")[1].split(" ")[3]
        op = s[2].split(":")[1].split(" ")[4]
        o2 = s[2].split(":")[1].split(" ")[5]
        test = int(s[3].split(" ")[-1])
        t = int(s[4].split(" ")[-1])
        f = int(s[5].split(" ")[-1])
        l = None
        if op == "+":
            l = lambda x,y : x+y
        if op == "*":
            l = lambda x,y : x*y
        opList = [o1,o2]
        testset = Testset(test,t,f)
        ret = Monkey(itms,no,testset,l,opList)
        return ret






```

```python
mList = []
for x in ll:
    m = Monkey.fromList(x)
    mList.append(m)

for x in mList:
    x.testset.retT = mList[x.testset.t]
    x.testset.retF = mList[x.testset.f]
for _ in range(0,20):    
    for m in mList:
        m.inspect()
    #for m in mList:
    #    print(m.no,m.getList())
    #print("---")
#print(len(mList[0].items))
mList.sort()
print(mList[-1].inspeted*mList[-2].inspeted)
```
