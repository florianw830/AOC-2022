```python
from collections import defaultdict

f = open("input.txt", "r")
l = f.read().split("\n")[:-1]
oobY = 9999
maxY = 0
cave = defaultdict(int)
ll = [[(int(y.split(",")[0]),int(y.split(",")[1])) for y in x.split(" -> ")] for x in l]


def fillWalls(start,end,dct):
    stpX = -1 if end[0]-start[0] <0 else 1
    stpY = -1 if end[1]-start[1] <0 else 1
    for x in range(start[0],end[0]+stpX,stpX):
        for y in range(start[1],end[1]+stpY,stpY):

            pt = str(x) + "," + str(y)
            #print(pt)
            dct[pt] = 1
            
            #print(pt)
            
def checkSand(pt,x,y):
    _pt = (pt[0]+x,pt[1]+y)
    tar = f"{_pt[0]},{_pt[1]}"

    return cave[tar] == 0,_pt,tar

            
def fillSand(pt,cave):
    x = 0
    y = 1
    tar = f"{pt[0]},{pt[1]}"
    endTar = tar
    while True:
 
        if pt[1]+1 == maxY:
            
            cave[f"{pt[0]+1},{pt[1]+1}"] = 1
            cave[f"{pt[0]-1},{pt[1]+1}"] = 1
            cave[f"{pt[0]},{pt[1]+1}"] = 1
            
        #print(pt)
        #if pt[1] > oobY:
            #return False
        
        #DOWN
        endTar = tar
        x = 0
        y = 1
        cont = False
        cont, _pt,tar =checkSand(pt,x,y)
        if cont:
            pt = _pt
            continue
        #DOWN + LEFT
        x = -1
        cont, _pt,tar =checkSand(pt,x,y)
        if cont:
            pt = _pt
            continue
        #DOWN+RIGTH
        x = 1
        cont, _pt,tar =checkSand(pt,x,y)
        if cont:
            pt = _pt
            continue
        break
    cave[endTar] =2
    if pt[0] == 500 and pt[1] == 0:
        return False
    #print("end",endTar)
    
    return True
    

for x in ll:
    start = x[0]
    
    for i in range(1,len(x)):
        if start[1] > maxY:
            maxY = start[1]
        end = x[i]
        #print(start,end)
        fillWalls(start,end,cave)
        #print("---")
        start = end
maxY+=2
print(maxY)
ct = 0
while fillSand((500,0),cave):
    ct+=1
print(ct+1)
```
