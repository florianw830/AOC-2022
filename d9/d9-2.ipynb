```python
class MyElem:
    def __init__(self, x,y,name,tail=None):
        self.tail = tail
        self.x = x
        self.y = y
        self.clist = []
        self.name = name
    
    def setCoord(self,c):
        tt = [self.x,self.y]
        self.x =c[0]

        self.y = c[1]

        if self.tail != None:
            self.tail.setCoord(tt)


    def __str__(self) -> str:
        return f"({self.x},{self.y})"
    def move(self,tar,commit =True):
        f = tar[0]
        m = [0,0]
     
        if f == "R":
            m[0] = 1
        if f == "L":
            m[0] = -1
        if f == "U":
            m[1] = -1
        if f == "D":
            m[1] = 1
        if f == "UR":
            m[0] = 1
            m[1] = -1
        if f == "UL":
            m[0] = -1
            m[1] = -1
        if f == "DR":
            m[0] = 1
            m[1] = 1
        if f == "DL":
            m[0] = -1
            m[1] = 1
        
        for i in range(0,int(tar[1])):
            self.x += m[0]
            self.y += m[1]

            self.clist.append((self.x,self.y))

            if self.tail != None:
                tar1 = ["",""]
                
                if abs(self.y - self.tail.y) + abs(self.x - self.tail.x) >= 3:
                    if self.y - self.tail.y <0:
                        tar1[0] = "U"
                    else: 
                        tar1[0] = "D"
                    tar1[1] = "1"

                    if self.x - self.tail.x <0:
                        #print(tar1)
                        tar1[0] = tar1[0]+"L"
                    else: 
                        #print(tar1)
                        tar1[0] = tar1[0]+"R"
             
                elif abs(self.y - self.tail.y) == 2 and abs(self.x - self.tail.x) == 0:        
                    if self.y - self.tail.y <0:
                        tar1[0] = "U"
                    else: 
                        tar1[0] = "D"
                    tar1[1] = "1" 
  
                elif abs(self.y - self.tail.y) == 0 and abs(self.x - self.tail.x) == 2:        
                    if self.x - self.tail.x <0:
                        tar1[0] = "L"
                    else: 
                        tar1[0] = "R"
                    tar1[1] = "1"

                if tar1 != ["",""]:
                    self.tail.move(tar1)

                
              
    
```
